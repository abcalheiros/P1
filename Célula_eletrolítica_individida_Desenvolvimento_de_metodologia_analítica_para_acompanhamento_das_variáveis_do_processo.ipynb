# Barplot (Comparativo entre massas teóricas x reais)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x = [30.005,9.826,0.805,0.263]
y = ["Massa teórica", "Massa real","Massa teórica", "Massa real"]
z = ["NaClO", "NaClO","H2", "H2"]

df = pd.DataFrame({"Massa (g)":x," ":y, "class":z})
df.head()
```

```python
sns.barplot(x = ' ', y = 'Massa (g)', data = df, hue = "class", palette = "spring")
plt.show()
```

# Lineplot (Produção de massas no tempo)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel("P1.xlsx")
df.head()
```

```python
sns.lineplot(x= "Tempo real (s)", y = "MNaClO (g)", data = df, color = 'orange')
plt.show()
```

```python
sns.lineplot(x= "Tempo real (s)", y = "MH2 (g)", data = df, color = 'yellow')
plt.show()
```

# Scatterplot (Mostrar correlação entre variáveis)

```python
sns.scatterplot(x = " Corrente (A)", y = "MNaClO (g)", data = df,alpha = 0.2, color = "black")
plt.show()
```

```python
#Massa NaClO gerada (g)
soma_massana = df['MNaClO (g)'].sum()
soma_massana
```

```python
#Massa H2 gerada (g)
soma_massah2 = df['MH2 (g)'].sum()
soma_massah2
```

```python
#Corrente média (A)
corrente_media = df[' Corrente (A)'].mean()
corrente_media
```

# PLANEJAMENTO EXPERIMENTAL

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel("Planilha (2).xlsx")
df.head()
```

```python
sns.barplot(x = "F1 (I)", y = 'Massa real NaClO (mg)', data = df, facecolor = [1,1,1,0], edgecolor = '0.2', ci = False)
plt.show()
```

```python
sns.barplot(x = "F2 (CNaCl)", y = 'Massa real NaClO (mg)', data = df, facecolor = [1,1,1,0], edgecolor = '0.2', ci = False)
plt.show()
```

```python
sns.barplot(x = "F3 (Salmoura)", y = 'Massa real NaClO (mg)', data = df, facecolor = [1,1,1,0], edgecolor = '0.2', ci = False)
plt.show()
```

```python
sns.barplot(x = "F1 (I)", y = 'Massa real NaClO (mg)', data = df, facecolor = [1,1,1,0], edgecolor = '0.2')
sns.barplot(x = "F2 (CNaCl)", y = 'Massa real NaClO (mg)', data = df, facecolor = [1,1,1,0], edgecolor = '0.2')
sns.barplot(x = "F3 (Salmoura)", y = 'Massa real NaClO (mg)', data = df, facecolor = [1,1,1,0], edgecolor = '0.2')
plt.show()
```

```python
sns.barplot(x = "EXPERIMENTO", y = 'Massa real NaClO (mg)', data = df, hue = "F1 (I)", palette = "Pastel1")
plt.show()
```
